```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/student-depression-dataset/Student Depression Dataset.csv")

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Academic Pressure"], y=df["Study Satisfaction"], hue=df["Depression"], palette={0: "blue", 1: "red"})

# Labels and title
plt.xlabel("Academic Pressure")
plt.ylabel("Study Satisfaction")
plt.title("Academic Pressure vs Study Satisfaction")

# Show plot
plt.legend(title="Depression", labels=["No Depression", "Depression"])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/student-depression-dataset/Student Depression Dataset.csv")

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df["Family History of Mental Illness"],
    y=df["Have you ever had suicidal thoughts ?"],
    hue=df["Depression"],
    palette={0: "blue", 1: "red"}
)

# Labels and title
plt.xlabel("Family History of Mental Illness")
plt.ylabel("Suicidal Thoughts")
plt.title("Family History of Mental Illness vs Suicidal Thoughts")

# Show plot
plt.legend(title="Depression", labels=["No Depression", "Depression"])
plt.show()
```
