```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = "/kaggle/input/student-depression-dataset/Student Depression Dataset.csv"
df = pd.read_csv(file_path)

# Group data by Dietary Habits and Depression status
dietary_depression_counts = df.groupby(["Dietary Habits", "Depression"]).size().unstack(fill_value=0)

# Categories and values
categories = dietary_depression_counts.index
depression_yes = dietary_depression_counts[1]  # Individuals with depression
depression_no = dietary_depression_counts[0]   # Individuals without depression

# Bar plot setup
bar_width = 0.35
x = np.arange(len(categories))

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for depression and no depression groups
ax.bar(x - bar_width/2, depression_yes, bar_width, label='Depression', color='b')
ax.bar(x + bar_width/2, depression_no, bar_width, label='No Depression', color='g')

# Labels and titles
ax.set_xlabel('Dietary Habits')
ax.set_ylabel('Number of Individuals')
ax.set_title('Depression vs Dietary Habits')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45)
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Group data by Profession and Depression status
counts = df.groupby(["Profession", "Depression"]).size().unstack(fill_value=0)

# Plot setup
x = np.arange(len(counts))
bar_width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - bar_width/2, counts[1], bar_width, label='Depression', color='r')
plt.bar(x + bar_width/2, counts[0], bar_width, label='No Depression', color='b')

# Labels and formatting
plt.xlabel('Profession')
plt.ylabel('Number of Individuals')
plt.title('Depression vs Profession')
plt.xticks(x, counts.index, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/kaggle/input/student-depression-dataset/Student Depression Dataset.csv")

# Create scatter plot with hue based on depression status
plt.figure(figsize=(10, 5))
sns.histplot(df, x="CGPA", hue="Depression", bins=30, kde=True, palette={0: 'blue', 1: 'red'}, alpha=0.5)

# Labels and title
plt.xlabel('CGPA')
plt.ylabel('Count')
plt.title('CGPA Distribution Based on Depression Status')

# Show plot
plt.legend(title="Depression", labels=["No Depression", "Depression"])
plt.show()
```
